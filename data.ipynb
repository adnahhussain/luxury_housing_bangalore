```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(r"C:\Users\Welcome\Desktop\luxury_housing_bangalore\Luxury_Housing_Bangalore.csv")
print(data)
```

```python
data.head()      
data.shape        
data.info()       
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.columns = data.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('-', '_')
data.columns
```

```python
for col in data.columns:
    if data[col].dtype in ['float64', 'int64']:
        data[col] = data[col].fillna(data[col].mean())
    else:
        data[col] = data[col].fillna('Unknown')

```

```python
data= data.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
for col in data.columns:
    if 'price' in col or 'cost' in col:
     data[col] = pd.to_numeric(data[col], errors='coerce')

for col in data.columns:
    if 'date' in col:
     data[col] = pd.to_datetime(data[col], errors='coerce')
```

```python
data.head()      
data.shape        
data.info()       
```

```python
clean_path = r"C:\Users\Welcome\Desktop\luxury_housing_bangalore\Luxury_Housing_Bangalore_Cleaned.csv"
data.to_csv(clean_path, index=False)
print("Cleaned data saved at:", clean_path)
```
